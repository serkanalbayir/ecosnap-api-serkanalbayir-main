```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
```

```python
#Define coordinates of where we want to center our map
boulder_coords = [40.015, -105.2705]

#Create the map
my_map = folium.Map(location = boulder_coords, zoom_start = 13)

#Display the map
my_map
```

```python
#Define the coordinates we want our markers to be at
Visually_impaired_path = [39.8656764, 32.7338174]
Weak_Wifi = [39.8656348, 32.7338507]
Electrical_Problem = [39.879809, 32.7133814]
Bad_Sidewalk = [39.880049, 32.7136179]
Other_Problems=[39.8712586, 32.7360446]

#Add markers to the map
folium.Marker(Visually_impaired_path, popup = 'Visually_impaired_path').add_to(my_map)
folium.Marker(Weak_Wifi, popup = 'Weak_Wifi').add_to(my_map)
folium.Marker(Electrical_Problem, popup = 'Electrical_Problem').add_to(my_map)
folium.Marker(Bad_Sidewalk, popup = 'Bad_Sidewal').add_to(my_map)
folium.Marker(Other_Problems, popup = 'Other_Problems').add_to(my_map)

#Display the map
my_map
```
